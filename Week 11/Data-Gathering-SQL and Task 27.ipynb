```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome()


phone_image = []
phone_name = []
average_ratings = []
total_rating = []
total_reviews = []
discount_price = []
actual_price = []

for i in range(1, 289):
    # Open the page
    driver.get(f"https://www.flipkart.com/search?q=smartphones&page={i}")
    
    # Give some time for the page to load
    time.sleep(3)
    
    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    # Extract the divs that contain phone data
    divs = soup.find_all('div', class_='tUxRFH')

    for div in divs:
        try:
            img = div.find('img', class_='DByuf4')
            phone_image.append(img['src'] if img else 'nan')
        except:
            phone_image.append('nan')

        try:
            name = div.find('div', class_='KzDlHZ')
            phone_name.append(name.text if name else 'nan')
        except:
            phone_name.append('nan')

        try:
            avg_rating = div.find('div', class_='XQDdHH')
            average_ratings.append(avg_rating.text if avg_rating else 0)
        except:
            average_ratings.append(0)

        try:
            total = div.find('span', text=lambda x: x and 'Ratings' in x)
            total_rating.append(total.text.split()[0] if total else 0)
        except:
            total_rating.append(0)

        try:
            review = div.find('span', text=lambda x: x and 'Reviews' in x)
            total_reviews.append(review.text.split()[0] if review else 0)
        except:
            total_reviews.append(0)

        try:
            discount = div.find('div', class_='Nx9bqj _4b5DiR')
            discount_price.append(discount.text if discount else 0)
        except:
            discount_price.append(0)  # Fixed typo

        try:
            actual = div.find('div', class_="yRaY8j ZYYwLA")
            actual_price.append(actual.text if actual else 0)
        except:
            actual_price.append(0)

# Quit the browser after scraping all pages
driver.quit()
```

```python
df = pd.DataFrame({'phone_name':phone_name,'average_ratings':average_ratings,'total_ratings':total_rating,'total_reviews':total_reviews,
             'discounted_price':discount_price,'actual_price':actual_price,'phone_image':phone_image})
```

```python
df['total_reviews'] = df['total_ratings'].apply(lambda x: x.split("&")[-1].strip() if "&" in x else None)
```

```python
df['total_ratings'] = df['total_ratings'].apply(lambda x: x.split("&")[1].strip() if "&" in x else None)
```
