```python
#  find number of movies starting with A for each group
def foo(group):
  return group[group['Series_Title'].str.startswith('A')]
```

```python
genres.apply(foo)
```

```python
# find ranking of each movie in the group according to IMDB score
def movieRank(group):
  group['ranking']=group['IMDB_Rating'].rank(ascending = False)
  return group
```

```python
genres.apply(movieRank)
```

```python
# find normalized IMDB rating group wise
def normalizedRating(group):
  group['Normalized_Rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group
```

```python
genres.apply(normalizedRating)
```

```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.head(2)
```

```python
ipl.shape
```

```python
ipl.info()
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(9)
```

```python
# find the batsman with max no of six
six = ipl[ipl['batsman_runs']==6]
six.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
over5 = ipl[(ipl['over']>15) & ((ipl['batsman_runs']==4) | (ipl['batsman_runs']==6))]
over5.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
# find V Kohli's record against all teams
temp = ipl[ipl['batsman']=='V Kohli']
temp.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# Create a function that can return the highest score of any batsman
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
# Create a function that can return the highest score of any batsman
def highestScore(batsman):
  tempDf=ipl[ipl['batsman']==batsman]
  return tempDf.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]

highestScore('V Kohli')
```
