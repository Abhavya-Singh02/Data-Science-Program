```python
import pandas as pd
import numpy as np
```

```python
# What are vectorized operations
#  vectorized operation is a operation that are used to do operation on vectors.
a=np.array([1,2,3,4]) #-> vector
a*4 # -> single vector operation on all array
```

```python
# problem in vectorized operations in vanilla python
# Ex:-
s = ['cat','mat',None,'rat']
# Q. in list which word are starting to the letter 'C'.
# we can use loop and list comprhension to solve this '
[i for i in s if i.startswith('c')] # -> this is the problem in vectorized operations.
```

```python
# How pandas solve this problem ?

s = pd.Series(['cat','mat',None,'rat'])

# string accessor
s.str.startswith('c')
# this is also fast and optimized for big datasets.
```

```python
# split -> get
df['lastName']=df['Name'].str.split(',').str.get(0)
```

```python
df[['title','firstName']]=df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
# this n tells to how many word can split
# expand function convert all the words in dataframe
```

```python
df['title'].value_counts()
```

```python
# replace
df['title']=df['title'].str.replace('Ms.','Miss.')
```

```python
df['title'].value_counts()
```

```python
# filtering
# startswith/endswith
df['firstName'][df['firstName'].str.startswith('A')]
```

```python
# Advanced filtering
# applying regex
# contains
# Q. search john -> both case
df['firstName'][df['firstName'].str.contains('John',case=False)]
```

```python
# find lastnames with start and end char vowel
#regex:-
# ^ - first char
#  . - any  char
# + - more than one char
# $ - last  char
# ^[^] - not first char
# * - allow multiple character
df['lastName'][df['lastName'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```

```python
df['lastName'][df['lastName'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```
